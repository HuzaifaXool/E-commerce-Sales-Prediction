```python
numerical_vals=data.select_dtypes(include=['float','int'])
```

```python
numerical_vals
```

```python
null_cols={}
for i in data.columns:
    if data[i].isna().sum()>0:
        null_cols[i] = {
            'Missing Values': data[i].isna().sum(),
            'Percentage': round((data[i].isna().mean() * 100), 2)
        }
        
```

```python
null_cols_df=pd.DataFrame.from_dict(null_cols, orient='index')
```

```python
null_cols_df.sort_values(by='Missing Values',ascending=False,inplace=True)
```

```python
null_cols_df
```

```python
threshold=10
cols_to_drop=null_cols_df[null_cols_df['Percentage']>threshold].index
```

```python
cols_to_drop
```

```python
data.drop(columns=cols_to_drop,inplace=True)
```

```python
data.shape
```

```python
cols_to_process=null_cols_df[null_cols_df['Percentage']<threshold]
```

```python
cols_to_process
```

```python
data.dropna(inplace=True)
```
